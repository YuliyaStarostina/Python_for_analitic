# Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Вывод: чем больше площадь недвижимости, тем цена выше

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

Вывод: рост цены напрямую зависит от качества недвижимости. Чем качественнее жилье, чем выше его класс, тем выше цена.

Вывод: Наибольшу цену имеют дома с максимальным грейдом

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Зависимость цены от этажности дома');
```

Вывод: Наибольшую стоимость имеют дома с этажностью 2.5

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('view')
plt.title('Зависимость цены от вида из окна');
```

Вывод: чем выше оценка вида из окна, тем выше стоимость недвижимости. В данном случае максимальную цену имеют дома, вид из которых оценивается на 4.

```python
sns.boxplot(x=df['price'], y=df['bedrooms'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('bedrooms')
plt.title('Влияние количества ванных комнат на цену');
```

Вывод: Количество ванных комнат окахывает влияние на цену, чем их больше, тем выше цена.
